```python
import random
import string

existed = ["gG"[(x>>3)&0b1]+"oO"[(x>>2)&0b1]+"oO"[(x>>1)&0b1]+"dD"[x&0b1] for x in range(16)]

generated = []
for x in range(10000):
    new = "".join(random.choices(string.ascii_letters, k=4))
    if new in existed:
        continue
    base = random.choice(existed)
    existed.append(new)
    generated.append((base,new))
```

```python
len(generated)
```

```python
with open("3.cs","w") as f:
    f.write("""namespace KoitoCoco.MoeCtf
{""")
    for base,new in generated:
        f.write(f"""
    public class FlagMachine_{new}: FlagMachine_{base}
    {{
        public override void SetFlag(dynamic flag) 
        {{ 
            base.SetFlag((byte[])flag);
            this.Flag ^= {random.randint(0, 0xffffffffffffffff)};
        }}
    }}
""")
    f.write("}")
```
